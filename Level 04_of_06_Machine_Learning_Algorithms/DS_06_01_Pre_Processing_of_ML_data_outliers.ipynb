# <span style="color:red"> Exercise 2: Delete Outliers</span>

```python
# simple function to remove outliers
def reject_outliers(data, m=2):
    return data[abs(data - np.mean(data)) < m * np.std(data)]
```

```python
without_outliers =reject_outliers(np_arr)
```

```python
np.count_nonzero(without_outliers)
```

```python
np.count_nonzero(np_arr)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(without_outliers)
plt.show()
```
